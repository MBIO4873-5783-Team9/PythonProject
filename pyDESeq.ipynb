```python
import pandas as pd
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats
```

```python
counts = pd.read_csv('C:/Users/15240/Desktop/OU/Python/project/counts1.csv')
counts
```

```python
counts = counts.set_index('GeneID')
counts
```

```python
counts = counts.T
```

```python
counts
```

```python
metadata = pd.DataFrame(zip(counts.index,['co','co','co','co','mono','mono','mono','mono']), columns = ['sample', 'condition' ])
metadata
```

```python
metadata = metadata.set_index('sample')
metadata
```

```python
dds = DeseqDataSet(counts = counts,
            metadata = metadata,
            design_factors = 'condition')
```

```python
dds
```

```python
dds.deseq2()
```

```python
dds
```

```python
stat_res = DeseqStats(dds, n_cpus = 8, contrast = ('condition', 'co','mono'))
```

```python
stat_res.summary()
```

```python
res = stat_res.results_df
```
